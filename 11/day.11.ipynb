### Part I
Simulate your seating area by applying the seating rules repeatedly until no seats change state. How many seats end up occupied?

```python
import numpy as np
import pandas as pd
```

```python
with open("input.sample") as f:
    lines = [list(line) for line in f.read().split("\n") if line]
board = np.array(lines)
board.shape
display(pd.DataFrame(board))
```

```python
%%time
def neighbours(x, y, board, value='#'):
    
    xmin = max(x-1, 0)
    ymin = max(y-1, 0)
    
    xmax = min(x+1, board.shape[0])
    ymax = min(y+1, board.shape[1])
    
    
    window = board[xmin:xmax+1, ymin:ymax+1]
    n = (window == value).sum()
    
    if board[x,y] == value:
        n -= 1
    return n
```

```python
def day11p1(infile, neighbours=neighbours, maxneigh=4):

    with open(infile) as f:
        lines = [list(line) for line in f.read().split("\n") if line]

    board = np.array(lines)
    #display(pd.DataFrame(board))
    
    round = 0
    while True:

        total = (board == "#").sum()

        new = board.copy()
     
        display(pd.DataFrame(board))
        
        for i in range(board.shape[0]):
            for j in range(board.shape[1]):

                if board[i, j] == ".":
                    continue
                
                n = neighbours(i, j, board)

                if board[i, j] == "L" and n == 0:
                    new[i, j] = "#"

                elif board[i, j] == "#" and n >= maxneigh:
                    new[i, j] = "L"

        
        new_total = (new == "#").sum()
        if new_total == total:
            return total

        #print(total)
        total = new_total
        board = new
        print(total)
        round += 1
        if round ==104:
            break
```

```python
%%time
day11p1("input.sample")
```

```python
%%time
day11p1("input")
```
