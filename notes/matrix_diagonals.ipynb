In linear algebra, the __main diagonal__ (sometimes principal diagonal, primary diagonal, leading diagonal, or major diagonal) of a matrix $A$  A is the collection of entries $A_{i, j}$ where $i = j$

The __antidiagonal__ (sometimes counter diagonal, secondary diagonal, trailing diagonal, minor diagonal, or bad diagonal) of a dimension $N$ square matrix, $B$ , is the collection of entries $B_{i,j}$ such that $i + j = N + 1$  for all $1 ≤ i , j ≤ N$ 

```python
import  numpy as np
import pandas as pd
m = np.arange(25).reshape(5,5)
pd.DataFrame(m)
```

```python
# Split the diagonal by the value
for i in range(m.shape[0]):
    for j in range(m.shape[0]):
        print(i, j)
        print(m[i, j])
        
        offset = j -i
        d = m.diagonal(offset)
        dj = max(0, offset)
        print(d)
        print(d[:j-dj], d[j+1-dj:])
        print()
        
```
